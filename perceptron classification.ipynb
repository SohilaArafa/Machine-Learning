```python
learning_rate=np.array([1,pow(10,-1),pow(10,-2),pow(10,-3),pow(10,-4),pow(10,-5),pow(10,-6),pow(10,-7),pow(10,-8),pow(10,-9)])
```

```python
def targetArray(x,targetnumber):
    target=np.zeros(2400)
    for i in range (2400):
        if (x[i]==targetnumber):
            target[i]=1
        else:
            target[i]=-1
    return target


target1=targetArray(train_labels,0) 
target2=targetArray(train_labels,1) 
target3=targetArray(train_labels,2) 
target4=targetArray(train_labels,3) 
target5=targetArray(train_labels,4) 
target6=targetArray(train_labels,5) 
target7=targetArray(train_labels,6) 
target8=targetArray(train_labels,7) 
target9=targetArray(train_labels,8) 
target10=targetArray(train_labels,9) 

```

```python
def perceptron(train,target,learning_rate,x):
    count=0
    w=np.zeros((785,),dtype=int)
    w[0]=1
    while (count<500):
        for j in range(2400):
            error=(w.T @train[j]) * target[j]
            if (error<0):
                #update missclassified point 
                w= w+(learning_rate[x]*train[j] *target[j])
                count+=1
                break
    #print(w)            
    #print(count)  
    return w 
```
